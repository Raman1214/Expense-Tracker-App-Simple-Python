## Expense Tracker Project

```python

expenseslist = []  # list of dictionaries

print("Welcome to Expenses Tracker App ")

while True:
    print("\n==== MENU ====")
    print("1. Add Expenses")
    print("2. View Expenses")
    print("3. View Total Expense")
    print("4. Exit")

    choice = input("Please Enter Your Choice: ")

    # 1. ADD EXPENSE
    if choice == "1":
        print("Add Your Expense:-------------------------")
        date = input("Enter the date ")
        category = input("Enter category (Food, Travel, Makeup, Books): ")
        description = input("Enter description: ")
        amount = float(input("Enter the amount: "))

        expense = {
            "date": date,
            "category": category,
            "description": description,
            "amount": amount,
        }

        expenseslist.append(expense)
        print("\nExpense is added successfully")

    # 2. VIEW ALL EXPENSES
    elif choice == "2":
        if len(expenseslist) == 0:
            print("No Expenses Added.")
        else:
            print("\n==== All Expense ====")
            count = 1
            for abc in expenseslist:
                print(
                    f"Expense {count}  -> {abc['date']}, "
                    f"{abc['category']}, "
                    f"{abc['description']}, "
                    f"{abc['amount']}"
                )
                count += 1
                

    # 3. VIEW TOTAL SPENDING
    elif choice == "3":
        total = 0
        for abc in expenseslist:
            total += abc["amount"]# total = total + abc["amount"]

        print("\nTotal Expense =", total)

    # 4. EXIT
    elif choice == "4":
        print("Dhanyavad🙏")
        break

    else:
        print("INVALID CHOICE. TRY AGAIN") 
```
